```python
import numpy as np
import pandas as pd
```

```python
X = pd.read_csv("/Users/ishaan/data-analysis-with-technical-indicators/Pair trading/Returns data/standardised_returns_for_clustering.csv")
X
```

```python
curr = X.columns[0]
X.rename(columns={curr: 'Stocks'}, inplace=True)
X = X.set_index('Stocks')
X

from sklearn.cluster import AffinityPropagation
ap = AffinityPropagation()
ap.fit(X)
labels = ap.predict(X)
```

```python
print(len(labels))
labels.max()
```

```python
groups = []
for i in range(labels.max()+1):
    arr = []
    for j in range(len(labels)):
        if labels[j] == i:
            arr.append(X.index[j])
    groups.append(arr)
```
