```python
def first_last_diff(x):
    return ((x.iloc[-1] - x.iloc[0])/x.iloc[0])
weekly = data[data.columns].resample('W-FRI').agg(first_last_diff)
weekly
```

```python
def first_last_diff(x):
    return ((x.iloc[-1] - x.iloc[0]) / x.iloc[0])

# Resampling the data to monthly frequency and applying the first_last_diff function
monthly = data[data.columns].resample('M').agg(first_last_diff)
monthly
```

```python
monthly = monthly[:12]
monthly
```

```python
for i in monthly.columns:
    monthly[i] = (1 + monthly[i]).cumprod() - 1

monthly
```

```python
transposed=monthly.T
transposed=transposed.iloc[1:]
transposed
```
