```python
data['Date'] = pd.to_datetime(data['Date'])
data.dtypes
```

```python
data.set_index('Date', inplace=True)
data
```

```python
data.columns
```

```python
def first_last_diff(x):
    return ((x.iloc[-1] - x.iloc[0])/x.iloc[0])
weekly = data[data.columns].resample('W-FRI').agg(first_last_diff)
weekly
```

```python
weekly = weekly[:49]
weekly
```

```python
for i in weekly.columns:
    weekly[i] = (1 + weekly[i]).cumprod() - 1

weekly
```
